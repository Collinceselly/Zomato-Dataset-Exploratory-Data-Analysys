```python
df_country_code = pd.read_excel('datasets/2024/Zomato_Country_Code.xlsx')
df_country_code.head()
```

```python
## Combining the df and df_country_code datasets

final_df = pd.merge(df, df_country_code, on = 'Country Code', how = 'left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_values = final_df.Country.value_counts().values
```

```python
country_values
```

```python
## Pie Chart

plt.pie(country_values[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

```python
## Observation: The maximum Zomato transactions are recorded in India, followed by United States and then United Kingdom
```

```python
final_df.columns
```

```python
final_df.head()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', hue='Rating color', data=ratings, palette=['blue','red','orange','orange','yellow','green','green'])
```

```python
ratings.sort_values(by='Rating count', ascending = False)
```

```python
ratings_per_text = ratings[['Rating text', 'Rating count']]
ratings_per_text
```

```python
rating_text = ratings_per_text.groupby(['Rating text']).sum().sort_values(by='Rating count', ascending = False)
rating_text
```

```python
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating text', data=ratings, palette=['blue','red','orange','orange','yellow','green','green'])
```

```python
## Find the country names with zero ratings
```

```python
final_df.head()
```

```python
country_and_rating = final_df[['Country','Aggregate rating']]
country_and_rating
```

```python
condition = (country_and_rating['Aggregate rating'] == 0.0)
zero_rated = country_and_rating[condition]
zero_rated
```

```python
zero_rated.groupby('Country').size().reset_index().sort_values(by = (''))
```

```python
## Find out which currency is used by which country?
```

```python
final_df.columns
```

```python
country_currency = final_df[['Country','Currency']].groupby(['Country', 'Currency']).size().reset_index()
country_currency
```

```python
## Which countries has online deliveries?
```

```python
final_df.head()
```

```python
online_delivery = final_df[['Country', 'Has Online delivery']]
condition = (online_delivery['Has Online delivery'] == 'Yes')
countries_with_online_delivery = online_delivery[condition]
countries_with_online_delivery.groupby('Country').size().reset_index()
```

```python
no_online_delivery = final_df[['Country', 'Has Online delivery']]
condition = (online_delivery['Has Online delivery'] == 'No')
countries_with_no_online_delivery = no_online_delivery[condition]
countries_with_no_online_delivery.groupby('Country').size().reset_index()
```

```python
with_online_delivery = final_df[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
with_online_delivery
```
